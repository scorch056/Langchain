Sentence transformers on huggingface

```python
from langchain_huggingface import HuggingFaceEmbeddings
embedddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
text="This is a test document"
query_result = embedddings.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
doc_result = embedddings.embed_documents([text, "This is a real document"])
doc_result
```
